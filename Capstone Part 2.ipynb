```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import os
import seaborn as sns
import scipy
```

```python
%cd C:\Users\Zak White\My Datasets
    #input your working directory above

dfmulti = pd.read_csv('Salaries_Of_State_Agencies_-_Multi-Year.csv') 
    #data set downloaded from https://data.oregon.gov/Revenue-Expense/Salaries-Of-State-Agencies-Multi-Year/4cmg-5yp4/data
df2014 = pd.read_csv('Salaries__State_Agencies__As_of_June_30__2014.csv')
df2013 = pd.read_csv('state.empl.sal.asof.063013_csv.csv',header= 1) 
    #the headers are not read in correctly and added from the first row
df2012 = pd.read_csv('Salary__State_Agencies__As_of_June_30__2012.csv')
df2011 = pd.read_csv('Salaries__State_Agencies__Fiscal_Year_2011__Update_.csv')



```

```python
#Some data cleaning and rearraning of column names is needed to merge data.
# they are cleaned per year, but a lot of this code can be condensed and dispatched with.
dfmulti.rename(columns={'   AGENCY #':'AGENCY #'}, inplace=True)
df2014['FISCAL YEAR']='2014'
df2014['ANNUAL SALARY']=df2014['ANNUAL SALARY'].astype('str')
df2014['ANNUAL SALARY'] = [x.strip('$',) for x in df2014['ANNUAL SALARY']]
df2014['ANNUAL SALARY']=df2014['ANNUAL SALARY'].astype('float')

df2013['FISCAL YEAR']='2013'
df2013.rename(columns={' ANNUAL SALARY ':'ANNUAL SALARY'}, inplace=True)
#df2013.rename(columns={'ANNUAL SALARY':'ANNUAL_SALARY'}, inplace=True)
df2013['ANNUAL SALARY']=df2013['ANNUAL SALARY'].astype('str')
df2013['ANNUAL SALARY'] = [x.strip(' $') for x in df2013['ANNUAL SALARY']]
df2013['ANNUAL SALARY'] = [x.replace(',','') for x in df2013['ANNUAL SALARY']]
df2013['ANNUAL SALARY']=df2013['ANNUAL SALARY'].astype('float')

df2012['FISCAL YEAR']='2012'
df2012['ANNUAL SALARY']=df2012['ANNUAL SALARY'].astype('str')
df2012['ANNUAL SALARY'] = [x.strip('$') for x in df2012['ANNUAL SALARY']]
df2012['ANNUAL SALARY']=df2012['ANNUAL SALARY'].astype('float')

df2011['FISCAL YEAR']='2011'
df2011=df2011.drop(['Fiscal Year'],axis=1)
df2011['ANNUAL SALARY']=df2011['ANNUAL SALARY'].astype('str')
df2011['ANNUAL SALARY'] = [x.strip('$') for x in df2011['ANNUAL SALARY']]
df2011['ANNUAL SALARY']=df2011['ANNUAL SALARY'].astype('float')
```

```python
#combine dfs into a list to merge
dfs = [dfmulti,df2014,df2013,df2012,df2011]

df_final = pd.concat(dfs,sort=False)
df_final['ANNUAL SALARY'] = df_final['ANNUAL SALARY'].astype('float')
df_final['FISCAL YEAR'] = df_final['FISCAL YEAR'].astype('int')
df_final['CLASSIFICATION'] = [x.strip() for x in df_final['CLASSIFICATION']] #remove white spaces on classifications so they group better
df_final['AGENCY TITLE'] = [x.strip() for x in df_final['AGENCY TITLE']]
df_final.info()

```

```python
Positions = df_final.groupby(['CLASSIFICATION']).agg(['count', 'mean', 'std', 'min', 'median', 'max','sum']).drop(['AGENCY #','FISCAL YEAR'],axis=1)
Positions

#Let's take a brief look at all job classifications and see what jobs were needed the most over the last 6 years

```
